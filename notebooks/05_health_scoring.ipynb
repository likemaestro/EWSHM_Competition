# 05 — Health Scoring

**Status: implemented as notebook-backed scoring narrative**

This notebook is the presentation index for the final synthetic health-score interpretation. The score is derived from unsupervised reconstruction errors produced by the neural model workflow in notebook 04 and the v3 helper scripts.

## Current scoring workflow

- Train the reference model on SET1 baseline behavior.
- Keep the SET1-fitted normalization and trained model fixed.
- Run inference across every available AQUINAS SET.
- Aggregate per-channel reconstruction errors into per-event error scores.
- Use a SET1-derived percentile threshold to flag high-error events.
- Compare monthly error distributions, threshold exceedance rates, and temperature context to describe structural response stability.

The main implementation references are:

- `notebooks/azrmirz_fncs/evaluate_architecture_1_v3.py`
- `notebooks/azrmirz_fncs/inference_cross_set_v3.py`
- `notebooks/azrmirz_fncs/inspect_outlier_event.py`

## Health-score evidence

The cross-SET inference workflow produces per-event error CSVs, cross-SET histograms, boxplots, error-vs-temperature plots, summary tables, and ranked extreme events. These artifacts form the current structural health-score evidence: a stable baseline from SET1, monthly comparisons against that baseline, and a transparent threshold for anomalous response events.
