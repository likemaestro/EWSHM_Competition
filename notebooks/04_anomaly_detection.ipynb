# 04 — Anomaly Detection

**Status: implemented as notebook-backed analysis**

This notebook is the presentation index for the unsupervised anomaly-detection work. The reusable toolkit prepares aligned NN input tensors and deterministic split/normalization artifacts; the neural model experiments live in the notebook-side model helpers and scripts listed below.

The competition data has no labels, so the anomaly signal is derived from reconstruction error rather than supervised targets.

## Current approach

- Build event-aligned strain, ACC, and temperature tensors during preprocessing.
- Prepare reproducible train/validation/test splits with train-only normalization.
- Train attention-autoencoder variants on baseline data.
- Use reconstruction error as the unsupervised anomaly score.
- Inspect per-channel and per-event error distributions before cross-SET scoring.
